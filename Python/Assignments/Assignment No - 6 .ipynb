#### Q1 - Write a program to accept percentage from user and Display the grade according to the following criteria :


```python
marks = int(input("Enter Marks :- "))
if marks >= 90 :
    print("Grade-A")
elif marks >80 and marks<=90:
    print("Grade-B")
elif marks >=60 and marks <=80:
    print("Grade-C")
else:
    print("Grade-D")
    
```

#### Q2 - Write a program to accept the cost price of a bike and Display the road tax to be paid according to the following criteria :
-------->>> 

```python
cost_price = int(input("Enter cost price :- "))
if cost_price > 100000:
    tax = cost_price * 0.15
    print("Tax is :- ", tax)
elif cost_price > 50000 and cost_price <= 100000:
    tax = cost_price*0.10
    print("Tax is :- ",tax)
else:
    tax = cost_price*0.05
    print("Tax is :- ",tax)
```

#### Q4 - Check how many times a given number can be divided by 3 before it is less than or equal to 10.

```python
num = int(input("Eneter a number : "))
count = 0 
for i in range(1,num+1):
    if i%3 ==0:
        if count<10:
            count+=1
print(count)
    
```

#### Q5 - why and when to use while loop in python give a detailed descriprion with example.
------->>> A while loop is a control flow statement which allows code to be executed repeatedly, depending on whether a condition is satisfied or not.
Basically when programmer does not know how many times loop repeted that time while loop is used means if programmer not shore about the frequency of iteration that time while loop is used.

```python
# Example-->
num = int(input("Enter a number :- "))
temp = num
sum_1 = 0
while num > 0:
    sum_1=sum_1+num
    num = num - 1
print(f"The sum of {temp} natural number is :- {sum_1}")
```

#### Q6 - Use nested while loop to print 3 different pattern

```python
#-1->
num = int(input())
while num > 0 :
    j=5
    while j > 0:
        print("*",end=(" "))
        j=j-1
    print()
    num=num-1
```

```python
#-2->
num = int(input())
while num > 0 :
    j=num
    while j > 0:
        print("*",end=(" "))
        j=j-1
    print()
    num=num-1
```

```python
n = int(input('Enter number of rows : ')) 
i = 1
while i <= n :
    j = 1
    while j <= i:
        print("*", end = " ")
        j += 1
    print()
    i += 1
```
