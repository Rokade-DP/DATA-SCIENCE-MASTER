```python
data = [["name","email_id","numbers"],["sudh","sudh@gmail.com",77655],["krish","krish@gmail.com",98987886]]
```

```python
import csv
```

```python
with open("text3.csv",'w') as f:
    w = csv.writer(f)
    for i in data:
        w.writerow(i)
```

```python
with open("text3.csv",'r') as f:
    read = csv.reader(f)
    for i in read:
        print(i)
```

### Buffer Read And Wrrite & Other File Methods

```python
import io
```

```python
with open("text5.txt",'wb') as f:
    file = io.BufferedWriter(f)
    file.write(b"Lorem Ipsum is simply dummy text of the printing and typesetting industry. Lorem Ipsum has been the industry's standard dummy text ever since the 1500s, when an unknown printer took a galley of type and scrambled it to make a type specimen book. It has survived not only five centuries, but also the leap into electronic typesetting, remaining essentially unchanged. It was popularised in the 1960s with the release of Letraset sheets containing Lorem Ipsum passages, and more recently with desktop publishing software like Aldus PageMaker including versions of Lorem Ipsum.\n")
    file.write(b"This is my second line that i am trying to write")
    file.flush()
```

```python
logging.info("This is my info log")
```

```python
l = [1,2,3,4,5,[2,3,4],"jay","Rokade"]
```

```python
l1_int = []
l2_str = []
for i in l:
    logging.info("This is the start of my first for loop{}".format(l))
    logging.info("This is the value of i am logging{}".format(i))
    if type(i) == list:
        for j in i:
            logging.info("logging my j {j} and i is {i}".format(i = i, j=j))
            if type(j)==int:
                l1_int.append(j)
            elif type(i) == str:
                l2_str.append(j)
    elif type(i)==int:
        l1_int.append(i)
    elif type(i)==str:
        l2_str.append(i)
logging.info("This is my final result with all int {l1} and with all string {l2}".format(l1=l1_int,l2=l2_str))
        
print(l1_int)
print(l2_str)
                
```
