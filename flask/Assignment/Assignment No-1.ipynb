#### Q3. What is App routing in Flask? Why do we use app routes?
---->>>
- In Flask, app routing refers to the process of mapping URLs to specific functions within your application. This mapping is achieved using the @app.route() decorator, where app is an instance of the Flask application.

#### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.
--->>>
- In Flask, the url_for() function is used for URL building. It generates a URL to the given endpoint with the method provided. This function greatly simplifies the process of linking to specific routes within a Flask application, as it abstracts away the need to hardcode URLs.
